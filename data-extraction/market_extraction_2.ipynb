```python
import pandas as pd

daily_new = pd.read_csv('market_daily_new.csv')
# Assuming daily_new is your DataFrame
daily_new['date'] = pd.to_datetime(daily_new['date'])

# Set the 'date' column as the index
daily_new.set_index('date', inplace=True)

# Group by 'TICKER', resample to weekly frequency starting on Sunday,
# aggregate 'PRC' as the last value of the week, and 'VOL' as the sum of the week
weekly_new = daily_new.groupby('TICKER').resample('W-SUN').agg({'PRC': 'last', 'VOL': 'sum'}).reset_index()

# Adjust the 'date' to reflect the start of the week (subtracting 6 days)
weekly_new['date'] = weekly_new['date'] - pd.to_timedelta(6, unit='d')

# This code is prepared for a DataFrame like daily_new with 1.7 million rows, as per your description.
```

```python
weekly_new[weekly_new['TICKER'] == 'AAPL']
```

```python
weekly_market = weekly_new.copy()
```

```python
weekly_market['date'] = weekly_market['date'].apply(lambda x: x - pd.Timedelta(days=1))
weekly_market['year'] = weekly_market['date'].apply(lambda x: x.year)
weekly_market
```

```python
len(weekly_market.loc[(weekly_market['TICKER'] == 'MSFT') & (weekly_market['year'] == 2005)])
```

```python
unique_tickers = weekly_market['TICKER'].unique()
years = weekly_market['year'].unique()
error_log = pd.DataFrame(columns = ['ticker','year', 'number'])
for t in unique_tickers:
     for y in years:
          c = len(weekly_market.loc[(weekly_market['TICKER'] == t) & (weekly_market['year'] == y)])
          if c < 52:
               new_row = {'ticker': t,'year':y, 'number':c}     
               error_log.loc[len(error_log)] = new_row
```
