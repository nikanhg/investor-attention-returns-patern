Market Feature:
 PRC
 
Prc is the closing price or the negative bid/ask average for a trading day. If the closing price is not available on any given trading day, the number in the price field has a negative sign to indicate that it is a bid/ask average and not an actual closing price. Please note that in this field the negative sign is a symbol and that the value of the bid/ask average is not negative.

If neither closing price nor bid/ask average is available on a date, prc is set to zero. In a monthly database, prc is the price on the last trading date of the month. The price series begins the first month-end after the security begins trading and ends the last complete month of trading.

```python
# Examine negative prc
positive_prc = df2['PRC'] > 0

count_positive = positive_prc.value_counts()
count_positive
```

```python
# make negative PRC values to positive
df2['PRC'] = df2['PRC'].abs()
positive_prc = df2['PRC'] > 0

count_positive = positive_prc.value_counts()
count_positive
```

```python
# Examine the avaliable time period of companies
df2['year'] = df2['date'].str.split('-').str[0]

df2['year'] = df2['year'].astype(int)

result = df2.groupby('TICKER')['year'].nunique()

filtered_result = result[result < 19]

print(filtered_result)
```

```python
# extract vol, closing price 
df3 = df2[['TICKER','COMNAM','PRC','VOL','date']]
df3.head()
```

```python
# Weekly Basis
df4 = df3.copy()
df4.head()
```

```python
df4['date'] = pd.to_datetime(df4['date'])
df4['week_index'] = df4['date'].dt.strftime('%Y-%U')
df4.head(20)
```

```python
# Add new column - Weekly colsing price
df4['week_cls_prc'] = df4.groupby(['TICKER', 'week_index'])['PRC'].transform('last')
```

```python
df4.tail(50)
```

```python
# Add new column - Weekly Volume
df4['week_vol'] = df4.groupby(['TICKER', 'week_index'])['VOL'].transform('sum')
```

```python
df4.head(10)
```

```python
df5 = df4[['TICKER','COMNAM','week_index','week_cls_prc','week_vol']].drop_duplicates()
df5.head()
```

```python
# Add new column - Weekly Return = (Closing Price this week - Closing Price Last week) / Closing Price Last week
df5['weekly_return'] = (df5['week_cls_prc'] - df5.groupby('TICKER')['week_cls_prc'].shift(1)) / df5.groupby('TICKER')['week_cls_prc'].shift(1)

df5.head()
```

```python
# df 5 Weekly basis
df5.to_csv(r'C:\Users\yitin\Desktop\attention\investor-attention-returns-patern\data\Weekly_market.csv', index=False)
```

```python
# Quarterly basis
df6 = df3.copy()
df6['date'] = pd.to_datetime(df6['date'])
df6['year_quarter'] = df6['date'].dt.to_period('Q')

# Add new column - Quarterly colsing price
df6['quarter_cls_prc'] = df6.groupby(['TICKER', 'year_quarter'])['PRC'].transform('last')

# Add new column - Weekly Volume
df6['quarter_vol'] = df6.groupby(['TICKER', 'year_quarter'])['VOL'].transform('sum')

df6 = df6[['TICKER','COMNAM','year_quarter','quarter_cls_prc','quarter_vol']].drop_duplicates()
```

```python
# Add new column - Quarterly Return 
df6['quarterly_return'] = (df6['quarter_cls_prc'] - df6.groupby('TICKER')['quarter_cls_prc'].shift(1)) / df6.groupby('TICKER')['quarter_cls_prc'].shift(1)
df6.head()
```

```python
# df 6 Quarterly basis
df6.to_csv(r"C:\Users\yitin\Desktop\attention\investor-attention-returns-patern\data\Quarterly_market.csv", index=False)
```

```python
# Yearly basis
df7 = df3.copy()
df7['date'] = pd.to_datetime(df7['date'])
df7['year'] = df7['date'].dt.to_period('Y')

# Add new column - Yearly colsing price
df7['year_cls_prc'] = df7.groupby(['TICKER', 'year'])['PRC'].transform('last')

# Add new column - Yearly Volume
df7['year_vol'] = df7.groupby(['TICKER', 'year'])['VOL'].transform('sum')

df7 = df7[['TICKER','COMNAM','year','year_cls_prc','year_vol']].drop_duplicates()
```

```python
# Add new column - Yearly Return 
df7['yearly_return'] = (df7['year_cls_prc'] - df7.groupby('TICKER')['year_cls_prc'].shift(1)) / df7.groupby('TICKER')['year_cls_prc'].shift(1)
df7.head()
```
