```python
import pandas as pd
import numpy as np
```

```python
svi_df = pd.read_csv('../data/svi_edgar_weekly.csv')
market_df = pd.read_csv('../data/weekly_new.csv')
```

```python
svi_df
```

```python
print(svi_df['symbol'].nunique())
print(market_df['TICKER'].nunique())
```

```python
svi_df['date'] = pd.to_datetime(svi_df['date'])
market_df['date'] = pd.to_datetime(market_df['date'])
print(type(svi_df['date'][0]))
print(type(market_df['date'][0]))
```

```python
svi_df['day'] = svi_df['date'].apply(lambda x: x.day_name())
market_df['day'] = market_df['date'].apply(lambda x: x.day_name())
```

```python
print(svi_df['day'].unique())
print(market_df['day'].unique())
```

```python
market_df.rename(columns={'TICKER': 'symbol'}, inplace=True)
```

```python
df_final = pd.merge(svi_df, market_df, on=['date', 'symbol'], how='left')
```

```python
x = 0
for s in svi_df['symbol'].unique():
     if s in market_df['symbol'].unique():
          x += 1
print(x)
```

```python
df_final.rename(columns={'agregated_score': 'svi', 'amount':'edgar', 'PRC':'price', 'VOL': 'volume'}, inplace=True)
df_final = df_final[['symbol', 'date', 'svi', 'edgar', 'price', 'volume']]
df_final
```

```python
df_final.to_csv("../data/market_svi_edgar.csv")
```
