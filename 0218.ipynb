```python
for _, row in df_urls.iterrows():
    url = row['URL']
    date = row['Date']
    filename = f"/content/drive/MyDrive/0218/{date}.csv" #/content/drive/MyDrive/0218
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        # Use BytesIO to treat the binary content as a file-like object
        zip_file = ZipFile(BytesIO(response.content))

        # Extract the zip file. This step assumes there is only one CSV file in the zip
        zip_file.extractall("temp_dir")
        csv_filename = zip_file.namelist()[0]  # Assuming the first file is the CSV

        # Read the CSV file into a pandas DataFrame
        df = pd.read_csv(f"temp_dir/{csv_filename}")

        # Step 1: Extract CIK numbers from 'uri_path'
        df['extracted_CIK'] = df['uri_path'].str.extract('/data/(\d+)')

        # Convert extracted CIK to integer for matching
        df['extracted_CIK'] = pd.to_numeric(df['extracted_CIK'], errors='coerce')

        df = df.dropna(subset=["extracted_CIK"])
        # Step 2: Count occurrences of each CIK in 'uri_path'
        # Initialize a count column in df_final for storing the counts
        df_final['Count'] = 0

        # Iterate over each CIK in df_final to count its occurrences in df
        for index, row in df_final.iterrows():
            count = df[df['extracted_CIK'] == row['CIK']].shape[0]
            df_final.at[index, 'Count'] = count
        df_final.to_csv(f"{filename}")
        # Perform garbage collection manually
        gc.collect()
    else:
        print(f"Failed to download the file. Status code: {response.status_code}")
```

```python
# Replace 'your_file.txt' with the path to your text file
file_path = '/content/ticker.txt'

# For a space-separated file, you might also want to consider using 'delim_whitespace=True' if spaces are your delimiters
df_space = pd.read_csv(file_path, delim_whitespace=True)

file_path = '/content/cik-lookup-data.txt'

# Read the file into a list
with open(file_path, 'r', encoding='latin1') as file:
    lines = file.readlines()

# Create a DataFrame from the list
df_all = pd.DataFrame(lines, columns=['Data'])

# Optionally, strip newline characters if present
df_all['Data'] = df_all['Data'].str.strip()

file_path = '/content/companies_after_2005.csv'

df_company = pd.read_csv(file_path)

df_space['Symbol'] = df_space['Symbol'].str.upper()

df_merged = pd.merge(df_company, df_space, on="Symbol", how="left")

df_final = df_merged[['Name','Symbol', 'CIK']]

# Function to search and extract CIK
def search_and_assign_cik(name, df_all):
    for data in df_all['Data']:
        # Check if the name is in the data string
        if name.upper() in data.upper():
            # Extract the 10-digit CIK number
            cik = ''.join(filter(str.isdigit, data))
            return cik
    return pd.NA  # Return pd.NA if no match is found

# Iterate over df_final and assign CIK where a match is found
for index, row in df_final.iterrows():
  if pd.isna(row['CIK']):
    df_final.at[index, 'CIK'] = search_and_assign_cik(row['Name'], df_all)
```
